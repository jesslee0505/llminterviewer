```python
def get_top_problems_by_topic(topic, limit=10):
    """
    Get the highest frequency problems for a specific topic
    """
    query = f"""
    SELECT 
        ID, 
        TITLE, 
        DIFFICULTY, 
        FREQUENCY,
        URL
    FROM 
        LC
    WHERE 
        RELATED_TOPICS LIKE '%{topic}%'
    ORDER BY 
        FREQUENCY DESC
    LIMIT {limit}
    """
    
    cursor.execute(query)
    results = cursor.fetchall()
    
    columns = ['id', 'title', 'difficulty', 'frequency', 'url']
    return pd.DataFrame(results, columns=columns)

# Example usage - Get high frequency Dynamic Programming problems
dp_problems = get_top_problems_by_topic('Dynamic Programming')
print("High frequency Dynamic Programming problems:")
print(dp_problems)

# Other common topics
array_problems = get_top_problems_by_topic('Array')
print("\nHigh frequency Array problems:")
print(array_problems)
```

```python
# 1. Query high frequency problems by company
def get_top_problems_by_company(company, limit=10):
    """
    Get high frequency problems commonly asked by a specific company
    """
    query = f"""
    SELECT 
        ID, 
        TITLE, 
        DIFFICULTY, 
        FREQUENCY,
        URL,
        RELATED_TOPICS
    FROM 
        LC
    WHERE 
        COMPANIES LIKE '%{company}%'
    ORDER BY 
        FREQUENCY DESC
    LIMIT {limit}
    """
    
    cursor.execute(query)
    results = cursor.fetchall()
    
    columns = ['id', 'title', 'difficulty', 'frequency', 'url', 'topics']
    return pd.DataFrame(results, columns=columns)

# 2. Query by difficulty and topic
def get_problems_by_difficulty_and_topic(difficulty, topic, limit=10):
    """
    Get high frequency problems with specified difficulty and topic
    """
    query = f"""
    SELECT 
        ID, 
        TITLE, 
        DIFFICULTY, 
        FREQUENCY,
        URL
    FROM 
        LC
    WHERE 
        DIFFICULTY = '{difficulty}'
        AND RELATED_TOPICS LIKE '%{topic}%'
    ORDER BY 
        FREQUENCY DESC
    LIMIT {limit}
    """
    
    cursor.execute(query)
    results = cursor.fetchall()
    
    columns = ['id', 'title', 'difficulty', 'frequency', 'url']
    return pd.DataFrame(results, columns=columns)

# 3. Get problems commonly asked by FAANG companies
def get_faang_problems(limit=20):
    """
    Get high frequency problems commonly asked by FAANG companies
    (Facebook, Amazon, Apple, Netflix, Google)
    """
    query = f"""
    SELECT 
        ID, 
        TITLE, 
        DIFFICULTY, 
        FREQUENCY,
        URL,
        COMPANIES
    FROM 
        LC
    WHERE 
        (COMPANIES LIKE '%Facebook%' OR 
         COMPANIES LIKE '%Amazon%' OR 
         COMPANIES LIKE '%Apple%' OR 
         COMPANIES LIKE '%Netflix%' OR 
         COMPANIES LIKE '%Google%')
        AND ASKED_BY_FAANG = 1
    ORDER BY 
        FREQUENCY DESC
    LIMIT {limit}
    """
    
    cursor.execute(query)
    results = cursor.fetchall()
    
    columns = ['id', 'title', 'difficulty', 'frequency', 'url', 'companies']
    return pd.DataFrame(results, columns=columns)

# 4. Get similar problem recommendations
def get_similar_problems(problem_id):
    """
    Get problems similar to the specified problem ID
    """
    # First get the original problem information
    query = f"""
    SELECT 
        TITLE,
        SIMILAR_QUESTIONS
    FROM 
        LC
    WHERE 
        ID = {problem_id}
    """
    
    cursor.execute(query)
    result = cursor.fetchone()
    
    if not result or not result[1]:
        return f"No problem found with ID {problem_id} or no similar problems exist"
    
    title, similar_questions = result
    
    print(f"Problems similar to '{title}':")
    
    # Parse the similar questions string
    # Format is typically: [title1, link1, difficulty1], [title2, link2, difficulty2], ...
    similar_list = []
    if similar_questions:
        # Simple parsing, might need more complex logic in practice
        parts = similar_questions.split('],')
        for part in parts:
            if '[' in part and ',' in part:
                clean_part = part.replace('[', '').replace(']', '').strip()
                items = clean_part.split(',')
                if len(items) >= 3:
                    similar_list.append({
                        'title': items[0].strip(),
                        'url': items[1].strip(),
                        'difficulty': items[2].strip()
                    })
    
    return pd.DataFrame(similar_list)

# 5. Comprehensive search function
def search_leetcode_problems(keyword=None, difficulty=None, topic=None, company=None, min_frequency=None):
    """
    Comprehensive search for LeetCode problems
    """
    conditions = []
    
    if keyword:
        conditions.append(f"LOWER(TITLE) LIKE '%{keyword.lower()}%'")
    
    if difficulty:
        conditions.append(f"DIFFICULTY = '{difficulty}'")
    
    if topic:
        conditions.append(f"RELATED_TOPICS LIKE '%{topic}%'")
    
    if company:
        conditions.append(f"COMPANIES LIKE '%{company}%'")
    
    if min_frequency is not None:
        conditions.append(f"FREQUENCY >= {min_frequency}")
    
    where_clause = " AND ".join(conditions) if conditions else "1=1"
    
    query = f"""
    SELECT 
        ID, 
        TITLE, 
        DIFFICULTY, 
        FREQUENCY,
        URL,
        RELATED_TOPICS,
        COMPANIES
    FROM 
        LC
    WHERE 
        {where_clause}
    ORDER BY 
        FREQUENCY DESC
    LIMIT 20
    """
    
    cursor.execute(query)
    results = cursor.fetchall()
    
    columns = ['id', 'title', 'difficulty', 'frequency', 'url', 'topics', 'companies']
    return pd.DataFrame(results, columns=columns)
```

```python
# Example 1: Get high frequency problems commonly asked by Amazon
amazon_problems = get_top_problems_by_company('Amazon')
print("High frequency problems asked by Amazon:")
print(amazon_problems)

# Example 2: Get hard difficulty dynamic programming problems
hard_dp = get_problems_by_difficulty_and_topic('Hard', 'Dynamic Programming')
print("\nHard difficulty dynamic programming problems:")
print(hard_dp)

# Example 3: Get problems commonly asked by FAANG companies
faang_problems = get_faang_problems()
print("\nProblems commonly asked by FAANG companies:")
print(faang_problems)

# Example 4: Get problems similar to "Two Sum" (ID=1)
similar_to_two_sum = get_similar_problems(1)
print("\nProblems similar to Two Sum:")
print(similar_to_two_sum)

# Example 5: Comprehensive search - Find medium difficulty array problems commonly asked by Google
google_medium_array = search_leetcode_problems(
    difficulty='Medium', 
    topic='Array', 
    company='Google', 
    min_frequency=50.0
)
print("\nMedium difficulty array problems commonly asked by Google:")
print(google_medium_array)
```
